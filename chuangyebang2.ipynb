```python
import requests
from bs4 import BeautifulSoup

text = ''
for i in range(1,31):
    url = 'http://www.cyzone.cn/event/list-764-0-' + str(i) + '-0-0-0-0/'
    resp = requests.get(url)
    text += resp.text

bs = BeautifulSoup(text,'html5lib')
rows = bs.select('tr.table-plate3')
    
names = []
urls = []
rounds = []
investors = []
times = []
for row in rows:
    item = row.select('td.tp2 > span.tp2_tit > a')[0]
    name = item.text.strip()
    url = item.get('href').strip()
    item = row.select('td.tp-mean + td')[0]
    round = item.text.strip()
    investor = row.select('td.tp3')[0].text.strip()
    time = row.select('td')[-2].text
        
    names.append(name)
    urls.append(url)
    rounds.append(round)
    investors.append(investor)
    times.append(time)

dic = {'名字':names,'地址':urls,'轮次':rounds,'投资人':investors,'时间':times}

import pandas as pd
dic_pd = pd.DataFrame(dic)
```
