```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.17huo.com/list-man-0-50011123-0--2m0.html'

names = []
prices = []
infos = []


while url:
    resp = requests.get(url)
    bs = BeautifulSoup(resp.text,'html5lib')
    items = bs.select('ul.item > li')
    
    for item in items:
        name = item.select('p.item_title')[0].text.strip()
        price = item.select('div.item_info > span.txt_nowprice')[0].text
        item_url = item.select('a')[0].get('href').strip()
        
        names.append(name)
        prices.append(price)
        infos.append(item_url)
        
    next_page = bs.select('font.page_down')
    if next_page:
        item = bs.select('div.pagem.product_list_pager > a')
        url = item[-2].get('href').strip()
    else:
        url =None
        
dic = {'名字':names,'价格':prices,'详情':infos}

import pandas as pd

dic_pd = pd.DataFrame(dic)
```
