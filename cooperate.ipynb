```python
import requests
from bs4 import BeautifulSoup

    
url = 'http://www.china-railway.com.cn/xwdt/jrtt/index.html'
resp = requests.get(url)
text = resp.text.encode('ISO-8859-1').decode('utf-8')
bs = BeautifulSoup(resp.text,'html5lib')
contents = []

items = bs.select('li.cf > a')
for item in items:
    url_tmp = 'http://www.china-railway.com.cn/xwdt/jrtt/' + item.get('href')[2:]
    resp = requests.get(url_tmp)
    bs = BeautifulSoup(resp.text.encode('ISO-8859-1').decode('utf-8'),'html5lib')
    content = bs.select('div.noyau-cr-content')[0].text.strip()

    contents.append(content)

```
