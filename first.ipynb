```python
import requests
from bs4 import BeautifulSoup
html = ''
for i in range(1,3):
    url = 'http://www.cyzone.cn/event/list-764-0-'+str(i)+'-0-0-0-0/'
    resp = requests.get(url)
    html += resp.text

bs = BeautifulSoup(html,'html5lib')
rows = bs.select('tr.table-plate3')

names = []
name_urls = []
rounds = []
investors = []
dates = []

for row in rows:
    item = row.select('td.tp2 > span.tp2_tit > a' )[0]
    name = item.text.strip()
    name_url = item.get('href').strip()
    round = row.select('td.tp-mean + td')[0].text.strip()
    investor = row.select('td.tp3')[0].text.strip()
    date = row.select('td')[-2].text.strip()
    
    names.append(name)
    name_urls.append(name_url)
    rounds.append(round)
    investors.append(investor)
    dates.append(date)

dic = {'name':names,'urls':name_urls,'round':rounds,'investor':investors,'date':dates}

import pandas as pd

dic_pd = pd.DataFrame(dic)

dic_pd
```
