transform

```python
nvda = pd.read_csv('NVDA.csv',index_col = 0,parse_dates=['Date'])
```

```python
nvda.head()
```

```python
nvda['year'] = nvda.index.year
```

```python
nvda.head()
```

```python
nvda.drop(['year'],axis=1,inplace=True)
```

```python
nvda.head()
```

```python
key = lambda x:x.year
nvda.groupby(key).agg(['mean','std']).head()
```

```python
my_transform = lambda x:(x-x.mean())/x.std()
transform = nvda.groupby(key).transform(my_transform)
transform.head()
```

```python
%matplotlib inline
```

```python
my_compare = pd.DataFrame({
    'origin Adj Close':nvda['Adj Close'],
    'Transformed Adj Close':transform['Adj Close']
})
```

```python
my_compare.head()
```

```python
my_compare.plot()
```

```python
my_compare.plot(kind = 'bar')
```

```python
price_range = lambda x : x.max() - x.min()
nvda.groupby(key).transform(price_range).head()
```
